```python
import random
import subprocess
```

```python
studentcount = 19
probcount = 3
myseed = 111218
random.seed(myseed)
studentlabel = 'A'
studentlist = []
for i in range(studentcount):
    studenttriple = (studentlabel, random.randint(0,100), random.randint(0,100))
    studentlist.append(studenttriple)
    studentlabel = chr(ord(studentlabel) + 1)
studentlist
```

```python
pycode = "homework/HW-VSFS/vsfs.py -s {} -n {}"
pyexecute = "python2.7"
command = pyexecute + " " + pycode
```

```python
for label, rand1, rand2 in studentlist:
    print('===========================================================')
    print('----------------------------- {} ---------------------------'.format(label))
    print('===========================================================')
    fullcommand1 = command.format(rand1, probcount)
    result1 = subprocess.run(fullcommand1.split(), stdout=subprocess.PIPE)
    print(result1.stdout.decode('utf-8'))
    fullcommand2 = command.format(rand2, probcount) + ' -r'
    print('-----------------------------------------------------------\n')
    result2 = subprocess.run(fullcommand2.split(), stdout=subprocess.PIPE)
    print(result2.stdout.decode('utf-8'))
```

```python
pycode = "homework/HW-VSFS/vsfs.py -s {} -n {} -c"
pyexecute = "python2.7"
command = pyexecute + " " + pycode
```
